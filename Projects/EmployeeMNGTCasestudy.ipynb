Create a Employee Management Case Study
Which Should have

1) Add Employee

2) Remove Employee by id

3) Remove employee by name

4) Update employee by id

5) Update employee by name

6) display all

7) Display By ID

8) Exit

Catch :- All data must be saved in CSV and when user re run code if any data exists should be loaded automatically

```python
import pandas as pd

class EmployeeManagementSystem:
    def __init__(self):
        self.employees = pd.DataFrame(columns=['Name', 'Age', 'ID', 'Department', 'Salary'])

    def save_to_csv(self):
        self.employees.to_csv('employees.csv', index=False)

    def load_csv(self):
        try:
            self.employees = pd.read_csv('employees.csv')
        except FileNotFoundError:
            pass

    def add_employee(self):
        id = input("Enter ID: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        department = input("Enter Department: ")
        new_employee = pd.DataFrame([[name, age, id, department, 0]], columns=['Name', 'Age', 'ID', 'Department', 'Salary'])
        self.employees = pd.concat([self.employees, new_employee], ignore_index=True)
        self.save_to_csv()
        print("Employee added successfully.")

    def remove_employee_by_id(self):
        id = input("Enter ID to remove: ")
        self.employees = self.employees[self.employees['ID'] != id]
        self.save_to_csv()
        print("Employee removed successfully.")

    def remove_employee_by_name(self):
        name = input("Enter Name to remove: ")
        self.employees = self.employees[self.employees['Name'] != name]
        self.save_to_csv()
        print("Employee removed successfully.")

    def update_employee_by_id(self):
        id = input("Enter ID to update: ")
        mask = self.employees['ID'] == id
        if mask.any():
            self.employees.loc[mask, 'Name'] = input("Enter New Name: ")
            self.employees.loc[mask, 'Age'] = input("Enter New Age: ")
            self.employees.loc[mask, 'Department'] = input("Enter New Department: ")
            self.save_to_csv()
            print("Employee updated successfully.")
        else:
            print("Employee not found.")

    def update_employee_by_name(self):
        name = input("Enter Name to update: ")
        mask = self.employees['Name'] == name
        if mask.any():
            self.employees.loc[mask, 'ID'] = input("Enter New ID: ")
            self.employees.loc[mask, 'Age'] = input("Enter New Age: ")
            self.employees.loc[mask, 'Department'] = input("Enter New Department: ")
            self.save_to_csv()
            print("Employee updated successfully.")
        else:
            print("Employee not found.")

    def display_all(self):
        print(self.employees)

    def display_by_id(self):
        id = input("Enter ID to display: ")
        mask = self.employees['ID'] == id
        if mask.any():
            print(self.employees[mask])
        else:
            print("Employee not found.")

ems = EmployeeManagementSystem()
ems.load_csv()

while True:
    print("\nEmployee Management System\n")
    print("1) Add Employee")
    print("2) Remove Employee by ID")
    print("3) Remove Employee by Name")
    print("4) Update Employee by ID")
    print("5) Update Employee by Name")
    print("6) Display All Employees")
    print("7) Display Employee by ID")
    print("8) Exit")

    choice = input("\nEnter your choice: ")

    if choice == '1':
        ems.add_employee()
    elif choice == '2':
        ems.remove_employee_by_id()
    elif choice == '3':
        ems.remove_employee_by_name()
    elif choice == '4':
        ems.update_employee_by_id()
    elif choice == '5':
        ems.update_employee_by_name()
    elif choice == '6':
        ems.display_all()
    elif choice == '7':
        ems.display_by_id()
    elif choice == '8':
        break
    else:
        print("Invalid choice. Please try again.")
```
