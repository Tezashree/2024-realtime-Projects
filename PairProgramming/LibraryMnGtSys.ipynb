```python
class Book:
    def __init__(self, title, author, genre):
        self.title = title
        self.author = author
        self.genre = genre
        self.is_available = True

    def __str__(self):
        availability = "Available" if self.is_available else "Not available"
        return f"{self.title} by {self.author} ({self.genre}) - {availability}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, genre):
        self.books.append(Book(title, author, genre))

    def remove_book(self, title):
        self.books = [book for book in self.books if book.title.lower() != title.lower()]

    def search_books(self, search_term):
        search_term = search_term.lower()
        return [book for book in self.books
                if search_term in book.title.lower()
                or search_term in book.author.lower()
                or search_term in book.genre.lower()]

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower() and book.is_available:
                book.is_available = False
                return book
        return None

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower() and not book.is_available:
                book.is_available = True
                return book
        return None

class User:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

    def __str__(self):
        return f"User: {self.name}\nBorrowed Books: {', '.join([book.title for book in self.borrowed_books])}"

library = Library()
library.add_book("summer love", "subin bhattarai", "Romance ")
library.add_book("muna madan", "laxmi prasad devkota", "epic poet")
library.add_book("Shirishko Phool", "parijat", " Fiction")

user = User("Aastha")

print("Welcome to the Library Management System")
while True:
    print("\nMenu:")
    print("1. Search for a book")
    print("2. Borrow a book")
    print("3. Return a book")
    print("4. View borrowed books")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        search_term = input("Enter title, author, or genre to search: ")
        results = library.search_books(search_term)
        if results:
            for idx, book in enumerate(results, 1):
                print(f"{idx}. {book}")
        else:
            print("No books found.")
    elif choice == "2":
        title = input("Enter the title of the book to borrow: ")
        book = library.borrow_book(title)
        if book:
            print(f"Book '{book.title}' borrowed successfully.")
            user.borrowed_books.append(book)
        else:
            print("Book not available.")
    elif choice == "3":
        title = input("Enter the title of the book to return: ")
        book = library.return_book(title)
        if book:
            print(f"Book '{book.title}' returned successfully.")
            user.borrowed_books.remove(book)
        else:
            print("Book not found in your borrowed list.")
    elif choice == "4":
        print(user)
    elif choice == "5":
        break
    else:
        print("Invalid choice. Please try again.")

```
